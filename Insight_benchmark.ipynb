## Insight project: benchmark
### Random guess

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
#generate data
data = np.random.uniform(0, 1, 200)
data = (data > 0.75).astype(int)
accuracy = []
precision = []
recall = []
f1 = []
fpr = []
for i in range(1, 1000):
    pred = np.random.uniform(0, 1, 200)
    pred = (pred > i / 1000).astype(int)
    accuracy.append(accuracy_score(data, pred))
    precision.append(precision_score(data, pred))
    recall.append(recall_score(data, pred))
    f1.append(f1_score(data, pred))
    tn, fp, fn, tp = confusion_matrix(data, pred).ravel()
    fpr.append(fp / (fp + tn))
    


plt.subplots_adjust(left=0.1, bottom=0.1, right=2, top=2, wspace=0.3, hspace=0.3)
plt.subplot(231)
plt.title('accuracy')
plt.ylabel('score')
plt.xlabel('random guess')
plt.xlim(0,100)
plt.ylim(0,1)
plt.scatter(np.arange(1,1000) / 10, accuracy)
plt.subplot(232)
plt.title('precision')
plt.xlabel('random guess')
plt.xlim(0,100)
plt.ylim(0,1)
plt.scatter(np.arange(1,1000) / 10, precision)
plt.subplot(233)
plt.title('recall')
plt.ylabel('score')
plt.xlabel('random guess')
plt.xlim(0,100)
plt.ylim(0,1)
plt.scatter(np.arange(1, 1000) / 10, recall)
plt.subplot(234)
plt.title('f1_score')
plt.xlabel('random guess')
plt.xlim(0,100)
plt.ylim(0,1)
plt.scatter(np.arange(1,1000) / 10, f1)
plt.subplot(235)
plt.title('roc')
plt.ylabel('true positive rate')
plt.xlabel('false positive rate')
plt.xlim(0,1)
plt.ylim(0,1)
plt.scatter(fpr, recall)
plt.subplot(236)
plt.title('precision-recall')
plt.ylabel('precision')
plt.xlabel('recall')
plt.xlim(0,1)
plt.ylim(0,1)
plt.scatter(recall, precision)



```
